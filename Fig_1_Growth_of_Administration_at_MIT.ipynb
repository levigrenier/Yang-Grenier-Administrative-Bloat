## Fig_1_Growth_of_Administration_at_MIT

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
```

```python
# Load the data from the Excel file
file_path = "MIT Data.xlsx"
df = pd.read_excel(file_path)

# Setting the 'Unnamed: 0' column as the index and renaming it to 'Year'
df = df.rename(columns={'Unnamed: 0': 'Year'}).set_index('Year')

# Base year for comparison
base_year = 1985

# Columns of interest
columns_of_interest = ['Undergraduates', 'Graduate Students', 'Faculty', 'Administrative Staff']

# Calculating the percentage change since 1985 for each required column
df_percentage_change = (df[columns_of_interest] - df.loc[base_year, columns_of_interest]) / df.loc[base_year, columns_of_interest] * 100

# Plotting the percentage change since 1985 with different colors, markers, and line styles
plt.figure(figsize=(10, 6))

line_styles = ['-', '--', '-.', ':']
markers = ['o', 's', 'D', '^']
colors = ['blue', 'orange', 'green', 'red']

for i, column in enumerate(columns_of_interest):
    plt.plot(df_percentage_change.index, df_percentage_change[column], 
             label=column, linestyle=line_styles[i], marker=markers[i], color=colors[i])

# Adding title, labels, and legend
plt.xlabel('Year')
plt.ylabel('Percentage Change (%)')
plt.legend()
plt.grid(True)
plt.show()
```
